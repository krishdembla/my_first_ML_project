# **My First Machine Learning Project**

# **Loading the Data**

```python
import pandas as pd

dataframe = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
dataframe
```

# **Preparing the Data**

## Separating the data into X and Y

```python
y = dataframe['logS']
y
```

```python
X = dataframe.drop('logS', axis = 1)
X
```

## Data Splitting

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
X_train
```

```python
X_test
```

# **Building the ML Model**

## **Linear Regression**

### **Training the model**

```python
from sklearn.linear_model import LinearRegression

linReg = LinearRegression()
linReg.fit(X_train, y_train)
```

### **Applying the trained model to make predictions**

```python
yTrain_prediction = linReg.predict(X_train)
yTest_prediction = linReg.predict(X_test)
```

```python
yTrain_prediction
```

```python
yTest_prediction
```

### **Assessment of the model's performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

linReg_train_mse = mean_squared_error(y_train, yTrain_prediction)
linReg_train_r2 = r2_score(y_train, yTrain_prediction)

linReg_test_mse = mean_squared_error(y_test, yTest_prediction)
linReg_test_r2 = r2_score(y_test, yTest_prediction)
```

```python
print('Linear Regression MSE (Trained): ', linReg_train_mse)
print('Linear Regression r2 (Trained): ', linReg_train_r2)
print('Linear Regression MSE (Test): ', linReg_test_mse)
print('Linear Regression r2 (Test): ', linReg_test_r2)
```

```python
results = pd.DataFrame(['Linear Regression', linReg_train_mse, linReg_train_r2, linReg_test_mse, linReg_test_r2 ]).transpose()
results.columns = ['Method','Trained MSE', 'Trained R2','Test MSE','Test R2']
```

```python
results
```

## **Random Forest**

### **Training the model**

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth = 2, random_state = 100)
rf.fit(X_train, y_train)
```

### **Applying the trained model to make predictions**

```python
yTrain_prediction_rf = rf.predict(X_train)
yTest_prediction_rf = rf.predict(X_test)
```

### **Assessment of the model's performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, yTrain_prediction_rf)
rf_train_r2 = r2_score(y_train, yTrain_prediction_rf)

rf_test_mse = mean_squared_error(y_test, yTest_prediction_rf)
rf_test_r2 = r2_score(y_test, yTest_prediction_rf)
```

```python
results_rf = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2 ]).transpose()
results_rf.columns = ['Method','Trained MSE', 'Trained R2','Test MSE','Test R2']
results_rf
```

## **Comparsion of LR and RF**

```python
model_comparison = pd.concat([results, results_rf], axis = 0).reset_index(drop = True)
model_comparison
```
